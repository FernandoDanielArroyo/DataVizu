## Exercise : load the data from pickle file

```python
# Load from the pickle file eco2mix
df = pd.read_pickle("data/eco2mix.pkl")
df.sample(10) # Expected Results
```

```python
# accumulated energy production / consummption by sum
df.isnull().astype(int).groupby(df.index.year).sum()
```

```python
# how many value is null ? Compare to understand where and why there is NaN
tmp = (df.groupby(df.index, dropna=False).sum() == 0).astype(int)
tmp.groupby(tmp.index.year).sum()
```

```python
# drop with columns
(df.loc[df.index.year==2018]
   .drop(columns=['Pompage', 'P√©rim√®tre', 'Consommation', 'Ech. physiques'])
   .sum()
)
```

```python
# drop with axis = 1
(df.loc[df.index.year==2018]
   .drop(['Pompage', 'P√©rim√®tre', 'Consommation', 'Ech. physiques'], axis=1)
   .sum(axis=0)
)
```

```python
df_2018 = df.loc[df.index.year == 2018].iloc[:, 0:9]
df_2018 = df_2018.drop(columns=['Consommation'])
df_2018 = df_2018.groupby(df_2018.index).sum()
df_2018.head()
```

```python
from itertools import combinations, product
```

```python
for i, value in enumerate(df_2018.columns):
    print(i, value)
```

```python
fig = plt.figure(figsize=(20, 15))
nrows = 7
ncols = 2

for i, col in enumerate(df_2018.columns):
    ax = fig.add_subplot(nrows, ncols, i+1)
    # df_2018.loc[:, col].dropna().plot(ax=ax, title=col)
    ax.plot(df_2018.loc[:, col].dropna())
    ax.set_title(col)
```

##  Exercise -  Pie chart 

Choose a year in the data, and plot the share for each of the different energy sources.

_Going further : potential drawbacks of this kind of plot. See : https://www.surveygizmo.com/resources/blog/pie-chart-or-bar-graph/_

```python
# Choose a year in the data, and plot the share for each of the different energy sources.
toviz = df.loc['2018'].groupby('P√©rim√®tre').sum().sum()
toviz = toviz[toviz > 0]
toviz = toviz[toviz.index !='Consommation']
ax = toviz.plot.pie(
    title='Production par fili√®re : ann√©e 2018',
    autopct='%.0f%%',
    textprops={'color':"w", 'size':14})
plt.ylabel('2018')
plt.legend(bbox_to_anchor=(1, 1));
```

Use the same data used to produce the pie chart, but this time use an horizontal bar.

```python
# first way by
plt.barh(toviz.sort_values().index, toviz.sort_values().values);
```

## Exercise : Area chart

Explore data trends thanks to an area chart.

```python
toviz = df.groupby(df.index.year).sum()
toviz = toviz.loc[:, toviz.columns != 'Consommation']
toviz
```

```python
toviz = df.groupby(df.index.year).sum()
toviz = toviz.loc[:, toviz.columns != 'Consommation']
(toviz.T / toviz.sum(axis=1)).T
```

```python
(toviz.T / toviz.sum(axis=1)).T.plot(kind='area')
plt.legend(bbox_to_anchor=(1, 1));
```

```python
# Explore trends
toviz = df.groupby(df.index.year).sum()
toviz = toviz.loc[:, toviz.columns != 'Consommation']
toviz = (toviz.T / toviz.sum(axis=1)).T
toviz.plot(kind='area')
plt.legend(bbox_to_anchor=(1, 1))
```

## Exercise : Scatter Plot

Scatter plots are helpful to put in evidence relationships between two variables. Can you find a data relationships and make it visible with a plot ?

_Bonus: try seaborn joint plots_

```python
# Explore variables relationships with scatter plot
# hypothesis : when nuclear is decreasing, thermic or other energy increase

toviz = (df.iloc[::2]
           .dropna(axis=0, how="all") # to drop the row full of NaN
           .resample('W') # to reduce weekly the number of value
           .sum() # aggregate
)
plt.scatter(toviz['Nucl√©aire'], toviz['Thermique']);
```

```python
import seaborn as sns
```

```python
sns.jointplot(
    x=toviz['Nucl√©aire'],
    y=toviz['Thermique']
    );
```

```python
sns.pairplot(data=toviz);
```
