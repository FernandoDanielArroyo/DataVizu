## Exploring energy data 

```python
# display data sample
df_energy.sample(10)
```

```python
# Creating dataframe to visualize 
toviz = df_energy.drop(columns=['DateTime']).dropna()
toviz
```

```python
# Question1 Energies
# Comment varient les différentes grandeurs présentées dans ce jeu de données?
axes = toviz.plot(figsize=(12,7), subplots=True)
fig = axes[0].get_figure()
plt.suptitle('Evolution de différentes sources de consommation et productions énergétiques nationales', fontsize=20)
plt.legend(fontsize=12)
fig.text(0.95, 0.5, 'Source : https://www.statistiques.developpement-durable.gouv.fr')
fig.supxlabel("L'évolution du temps en années", fontsize=15)
fig.supylabel('Energie en kilowatts', fontsize=15)
plt.show()
```

```python
# Question2 Energies 
# Quelle évolution de la consommation ?
fig = plt.figure() 
ax_2 = fig.add_subplot()
ax_2 = df_energy['Consommation'].dropna().plot(figsize=(12,7))
ax_2.set_title("Consommation en fonction du temps", fontsize=24, fontweight="bold")
ax_2.set_xlabel("Date", fontweight="bold", size =18)
ax_2.set_ylabel("Consommation", fontweight="bold", size = 18)
plt.text(2, 0.5, 'Source : https://www.statistiques.developpement-durable.gouv.fr')
plt.show()
```
